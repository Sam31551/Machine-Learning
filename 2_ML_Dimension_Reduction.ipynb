```python
data = data.dropna(axis='columns')
data
```

## Remove vars with low variance

```python
non_cont_vars = data[['area', 'area_code']]
non_cont_vars
```

```python
numeric = data.iloc[:,2:]
numeric
```

```python
#Get all values with less than 10% variance as they are unlikely to affect target variable
var = numeric.var()
numeric = numeric.columns
variable = [ ]
for i in range(0,len(var)):
    if var[i]>=10:   #setting the threshold as 10%
       variable.append(numeric[i+1])
```

```python
len(variable)
```

```python
type(variable)
```

```python
data = data[variable]
data
```

```python
total_data = pd.concat([non_cont_vars, data], axis=1)
total_data
```

## Remove correlated features

```python
income = total_data['gross_income']
income
```

```python
feature_cols = data.drop(['gross_income'], axis=1)
feature_cols
```

```python
def remove_collinear_features(total_data, threshold):
   # Calculate the correlation matrix
    corr_matrix = total_data.corr()
    iters = range(len(corr_matrix.columns) - 1)
    drop_cols = []
    # Iterate through the correlation matrix and compare correlations
    for i in iters:
        for j in range(i+1):
            item = corr_matrix.iloc[j:(j+1), (i+1):(i+2)]
            col = item.columns
            row = item.index
            val = abs(item.values)

            # If correlation exceeds the threshold
            if val >= threshold:
                # Print the correlated features and the correlation value
                #print(col.values[0], "|", row.values[0], "|", round(val[0][0], 2))
                drop_cols.append(col.values[0])

    # Drop one of each pair of correlated columns
    drops = set(drop_cols)
    total_data = total_data.drop(columns=drops)
    print('Removed Columns {}'.format(drops))
    return total_data
```

```python
data = remove_collinear_features(feature_cols, 0.5)
data
```
